1. Highest total runs scored.

```python
import psycopg2
import pandas as pds
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:2009anshulab@localhost:5432/ipl');

conn = engine.connect();

#function for plot style
def style():
    sns.set_style("whitegrid");
    sns.despine();
    plt.figure(figsize=(8,6));

df1 = pds.read_sql("select batter,sum(total_run) as total_runs_scored from ball_by_ball group by batter order by sum(total_run) desc limit 20",conn);

print(df1);

style();
plt.title("Top 10 Batsman",fontsize=15);
sns.barplot(x="batter",y="total_runs_scored",data=df1,palette='Reds');
plt.xticks(rotation=90);
plt.show();
```

2. Highest wickets taker. 

```python
df2 = pds.read_sql("select bowler,sum(isWicketDelivery) as wickets_taken from ball_by_ball group by bowler order by sum(isWicketDelivery) desc limit 20",conn);
print(df2);

plt.title("Highest wickets taken");
sns.barplot(x="wickets_taken",y="bowler",data=df2);
plt.show();
```

3. Teams with highest and lowest scores.

```python
df3 = pds.read_sql("select battingteam, sum(total_run) as total_runs_by_team from ball_by_ball group by battingteam order by sum(total_run) desc",conn);
print(df3);

style();
plt.title("Total runs per team");
sns.barplot(x="battingteam",y="total_runs_by_team",data=df3,palette='Oranges');
sns.despine();
plt.xticks(rotation=90);
```

4. Average of all batsman and batsman with highest average.

```python
df4 = pds.read_sql("select batter,(sum(total_run)/count(distinct id)) as average from ball_by_ball group by batter order by average desc limit 30",conn);
print(df4);

style();
plt.title("Average of Batsman");
sns.barplot(x="average",y="batter",data=df4,palette='rocket');
```

5. Number of Matches played by each team.

```python
df5 = pds.read_sql("select team1 as Team,count(team1) as \"No. of matches played\" from matches group by team1;",conn);
print(df5);

style();
plt.title("Number of matches played by each team");
sns.barplot(x="No. of matches played",y="team",data=df5,palette='gnuplot');
```

6. Perfomance of V Kohli in the last 20 matches.

```python
df6 = pds.read_sql("select id,sum(batsman_run) as \"Total runs in match\" from ball_by_ball where batter = 'V Kohli' group by id order by id desc limit 20",conn);
print(df6);

style();
plt.title("Virat Kohli's performance in last 20 matches");
sns.lineplot(x="Total runs in match",y="id",data=df6);
```

7. Top 15 players with Man of the match awards.

```python
df7 = pds.read_sql("select player_of_match as Player, count(player_of_match) as award from matches group by player order by count(player_of_match) desc limit 15",conn);

print(df7);

plt.title("Players with highest Player of the match awards!");
sns.barplot(x="award",y="player",data=df7,palette='tab10_r');
plt.show();
```

8. 3000 runs club with highest strike rate.

```python
df8 = pds.read_sql("select batter,batsman_runs,strike_rate from "+
"(select batter, batsman_runs,((batsman_runs*1.0)/total_balls)*100 as strike_rate from "+
"(select batter,sum(batsman_run) as batsman_runs, count(batter) as total_balls "+
"from ball_by_ball group by batter) as x order by strike_rate desc) as b "+
"where batsman_runs>=3000 order by strike_rate desc limit 30;",conn);

print(df8);

style();
plt.title("3000 run club");
sns.lineplot(x="batter",y="strike_rate",data=df8);
plt.xticks(rotation=90);
```

9. Economy rates for bowlers who have bowled atleast 50 overs.

```python
df9 = pds.read_sql("select bowler,total_runs_conceeded/(total_balls*1.0) as economy_rate from "+
"(select bowler, count(bowler) as total_balls, sum(total_run) as total_runs_conceeded "+
"from ball_by_ball group by bowler) as x "+
"where total_balls > 300;",conn);

print(df9);

style();
plt.title("Economy rates");
sns.histplot(x="economy_rate",color='green',kde=True,data=df9,bins=80);
```

10. Number of wins for each team.

```python
df10 = pds.read_sql("select winning_team as team,count(winning_team) as wins from matches group by winning_team;",conn);
print(df10);

style();
plt.title("Number of wins per team");
sns.set_style("whitegrid");
sns.barplot(x="team",y="wins",data=df10,palette='mako');
plt.xticks(rotation=90);
```
